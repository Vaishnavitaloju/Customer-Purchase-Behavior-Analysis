```python
# Percentage of missing values in Review Rating
df["Review Rating"].isnull().mean()*100
```

```python
# Impute missing Review Ratings using the median value within each Category
df["Review Rating"] = df.groupby("Category")["Review Rating"] \
                        .transform(lambda x: x.fillna(x.median()))
```


The Review Rating column had approximately 0.95% missing values To preserve data completeness while maintaining realistic category-level behavior, missing ratings were imputed using the **median review rating within each product Category**. 

- Why median? It is robust to outliers and better represents typical ratings than the mean.  
- Why category-wise? Different categories may have different rating patterns, and using the category median preserves these patterns.  

This ensures the dataset is complete and suitable for accurate analysis of customer reviews and product performance.

### Creating Age Groups

To analyze customer behavior by age, the continuous **age** column is transformed into categorical groups. This allows us to compare metrics like revenue, purchase frequency, or subscription rates across different life stages.

Two approaches are applied:

1. **Fixed bins (`pd.cut`)**: Groups ages based on specific ranges, e.g., 0-30, 31-45, 46-60, 61+. Useful if we have business-defined age categories.  
2. **Quantile-based bins (`pd.qcut`)**: Splits the age column into four groups with approximately equal number of customers in each. This ensures balanced groups for statistical comparisons.

Both methods assign a new column `age_group` to the dataset.

```python
# Define a mapping from frequency labels to estimated number of days
freq_map = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
# Define a mapping from frequency labels to estimated number of days
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(freq_map)
```

To perform numerical analysis on how frequently customers make purchases, we convert the `frequency_of_purchases` column — which contains categorical time periods like "Monthly" or "Annually" — into a numerical format in days.

A mapping dictionary (`freq_map`) is defined where each frequency label is assigned an estimated number of days. This allows for easy comparison and aggregation of purchase behaviors over time.

```python
from sqlalchemy import create_engine

# Step 1: Connect to MySQL
# Replace placeholders with your actual details
username = "root"             # default MySQL user
password = "root"    # the password you set in MySQL Workbench
host = "localhost"            # if running locally
port = "3306"                 # default MySQL port
database = "retail_database"  # the database you created earlier

# Step 2: Create SQLAlchemy engine for MySQL connection
# Use 'mysql+pymysql' dialect for connection
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Step 3: Load DataFrame into MySQL
table_name = "customer_data"     # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```

### Uploading the DataFrame to MySQL

After cleaning and preprocessing the dataset, we connect to the MySQL database using SQLAlchemy and PyMySQL. The `customer_data` DataFrame is uploaded to the `retail_database` as a new table.

- **Engine setup:** Defines connection parameters (username, password, host, port, database).  
- **Data upload:** The `to_sql` function creates the table if it doesn't exist or replaces it if it does (`if_exists='replace'`).  

This allows us to run **SQL queries directly on the database** for analysis and reporting.

### SQLAlchemy `create_engine` for Database Connection

`create_engine` is a function from the SQLAlchemy library used to **establish a connection between Python and a SQL database**. 

It allows Python scripts or notebooks to:

- Execute SQL queries directly from Python.
- Read from and write to SQL tables (like our `customer_data` table).
- Support multiple databases with a unified interface (MySQL, PostgreSQL, SQLite, etc.).

In this project, we use it with **MySQL + PyMySQL** to upload the cleaned dataset and run queries for customer behavior analysis.
